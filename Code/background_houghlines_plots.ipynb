```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
xs = [3,7,11]
ys = [7,5,3]
x = np.linspace(-1,20,100)
y = -0.5*x+8.5

plt.plot(x, y, '-r', label='y=-0.5*x+8.5')
plt.title('Graph with cartesian coordinate system')
plt.scatter(xs, ys, color='blue', marker='o')
names = ["(3,7)","(7,5)","(11,3)"]
for i, txt in enumerate(names):
    ax.annotate(txt, (xs[i], ys[i]))
plt.xlim(0, 20)
plt.ylim(-1, 10)
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.axhline(0, color='black')
plt.legend(loc='upper right')
plt.grid()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
xs = [3,7,11]
ys = [7,5,3]
x = np.linspace(-1,20,100)
theta = 1.1071
rho = 7.6
y = ((-1*np.cos(theta)/np.sin(theta)*x)+rho/np.sin(theta))

plt.plot(x, y, '-r', label=r'$y=(-\frac{\cos \theta}{\sin \theta}x+\frac{\rho}{sin \theta}),\theta=1.1071, \rho=7.6$')
plt.title('Graph with polar coordinate system')
plt.scatter(xs, ys, color='blue', marker='o')
names = ["(3,7)","(7,5)","(11,3)"]
for i, txt in enumerate(names):
    ax.annotate(txt, (xs[i], ys[i]))
plt.xlim(0, 20)
plt.ylim(-1, 10)
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.axhline(0, color='black')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
